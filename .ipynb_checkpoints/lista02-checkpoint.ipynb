# Exerc√≠cio 4

## $r = 1$

```python
L_4 = 1.0  # m
c_4 = 0.2  # 300.0  # m/s
dx_4 = 0.01  # m
dt_4 = dx_4 / c_4
I_4 = int(L_4 / dx_4)

t_f_4 = 2 * L_4 / c_4

x0_4 = 0.3  # m
k_4 = 1000  # m^-2

x_4 = np.linspace(0, L_4, I_4)
y0_4 = list(np.exp(- k_4 * (x_4 - x0_4)**2))
y0_4[0] = 0.0
y0_4[-1] = 0.0

y_4 = ondas.propagate(
    y_0=y0_4,
    t_f=t_f_4,
    dx=dx_4,
    c=c_4,
    pe='solta',
    pd='solta'
)

N_4 = len(y_4)

fig_4, ax_4 = plt.subplots()

ax_4.set_xlim(0, 1)
ax_4.set_ylim(-1, 1)
ax_4.set_title('Propaga√ß√£o de ondas numa corda com extremidades livres \n ($r$ = 1, $c$ = 0.2 m/s)')
ax_4.set_xlabel('$x$ (m)')
ax_4.set_ylabel('$y$ (cm)')

line_4, = ax_4.plot(x_4, y_4[0])


def ondas_1(i):
    line_4.set_ydata(y_4[i])
    return line_4,


ani_4 = animation.FuncAnimation(
    fig=fig_4,
    func=ondas_1,
    frames=N_4,
    interval=dt_4 * 1000,
    blit=True,
    repeat=True
)

fig_4.show()
```
